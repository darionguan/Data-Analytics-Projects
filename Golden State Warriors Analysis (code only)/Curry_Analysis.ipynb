```python
# Import libraries
import pandas as pd
import numpy as np
import pandasql as psql
```

```python
# Read games
filename = '/Users/darionguan/Desktop/NBA Analytics/games.csv'
games = pd.read_csv(filename)

# Plus one for each season
games['SEASON'] = games['SEASON'] + 1

# Convert to datetime format
games['GAME_DATE_EST'] = pd.to_datetime(games['GAME_DATE_EST'], format='%Y-%m-%d')

games
```

```python
# Create dataframe for regular season games in 2015
fifteen_df = games.loc[((games['GAME_DATE_EST'] >= '2014-10-28') & 
                        (games['GAME_DATE_EST'] <= '2015-04-15'))]
fifteen_df = fifteen_df[['GAME_DATE_EST', 'GAME_ID']]
fifteen_df
```

```python
# Create dataframe for regular season games in 2016
sixteen_df = games.loc[((games['GAME_DATE_EST'] >= '2015-10-27') & 
                        (games['GAME_DATE_EST'] <= '2016-04-13'))]
sixteen_df = sixteen_df[['GAME_DATE_EST', 'GAME_ID']]
sixteen_df
```

```python
# Read games_details
filename = '/Users/darionguan/Desktop/NBA Analytics/games_details.csv'
curry = pd.read_csv(filename)

# Subset to Stephen Curry's rows
curry = curry.loc[(curry['PLAYER_NAME'] == 'Stephen Curry')]

# Subset to GAME_ID and FG3M
curry = curry[['GAME_ID', 'FG3M']]

# Convert floats to int
# games_details['FG3M'] = games_details['FG3M'].astype(int)

curry
```

```python
# Query
q = """
SELECT f.GAME_DATE_EST AS game_date_est,
        f.GAME_ID AS game_id,
        c.FG3M AS fg3m
FROM fifteen_df AS f
INNER JOIN curry AS c
    ON f.GAME_ID = c.GAME_ID
"""

curry_15_df = psql.sqldf(q)

# Convert to datetime format
curry_15_df['game_date_est'] = pd.to_datetime(curry_15_df['game_date_est'], format='%Y-%m-%d')
curry_15_df = curry_15_df.sort_values(by='game_date_est')

# Drop duplicates
curry_15_df = curry_15_df.drop_duplicates()

# Fill NaN (DNP) with 0s
curry_15_df['fg3m'] = curry_15_df['fg3m'].fillna(0)

# Cumulate 3s made
curry_15_df['Total3s'] = curry_15_df['fg3m'].cumsum()

# Convert Total3s to int
curry_15_df['Total3s'] = curry_15_df['Total3s'].astype(int)

# Create GAME_NUM
curry_15_df['GAMES'] = 1
curry_15_df['GAME_NUM'] = curry_15_df['GAMES'].cumsum()
curry_15_df = curry_15_df.drop(['GAMES'], axis=1)

# Create SEASON column
curry_15_df['SEASON'] = 2015

# Reset index
curry_15_df = curry_15_df.reset_index(drop=True)

curry_15_df

# Convert to csv
# curry_15_df.to_csv('curry.csv', index=False)
```

```python
# Query
q = """
SELECT s.GAME_DATE_EST AS game_date_est,
        s.GAME_ID AS game_id,
        c.FG3M AS fg3m
FROM sixteen_df AS s
INNER JOIN curry AS c
    ON s.GAME_ID = c.GAME_ID
"""

curry_16_df = psql.sqldf(q)

# Convert to datetime format
curry_16_df['game_date_est'] = pd.to_datetime(curry_16_df['game_date_est'], format='%Y-%m-%d')
curry_16_df = curry_16_df.sort_values(by='game_date_est')

# Drop duplicates
curry_16_df = curry_16_df.drop_duplicates()

# Fill NaN (DNP) with 0s
curry_16_df['fg3m'] = curry_16_df['fg3m'].fillna(0)

# Cumulate 3s made
curry_16_df['Total3s'] = curry_16_df['fg3m'].cumsum()

# Convert Total3s to int
curry_16_df['Total3s'] = curry_16_df['Total3s'].astype(int)

# Create GAME_NUM
curry_16_df['GAMES'] = 1
curry_16_df['GAME_NUM'] = curry_16_df['GAMES'].cumsum()
curry_16_df = curry_16_df.drop(['GAMES'], axis=1)

# Create SEASON column
curry_16_df['SEASON'] = 2016

# Reset index
curry_16_df = curry_16_df.reset_index(drop=True)

curry_16_df

# Convert to csv
# curry_16_df.to_csv('curry.csv', index=False)
```

```python
curry_df = pd.concat([curry_15_df, curry_16_df], axis = 0)
curry_df = curry_df.reset_index(drop=True)

# Convert to csv
curry_df.to_csv('curry.csv', index=False)
```
