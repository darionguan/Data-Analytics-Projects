```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
filename = '/Users/darionguan/Desktop/Presidential Prediction/biden_polls.csv'
biden_polls = pd.read_csv(filename).iloc[: , 1:]
biden_polls.head()
```

```python
filename = '/Users/darionguan/Desktop/Presidential Prediction/trump_polls.csv'
trump_polls = pd.read_csv(filename).iloc[: , 1:]
trump_polls.head()
```

```python
# Effect of income on Biden's polling
# Median nonfamily income is a significant predictor on Biden's polls

X = biden_polls.iloc[:, 2:7]
y = biden_polls.iloc[:, 1:2]

# Define the model
X = sm.add_constant(X)
model = sm.OLS(y, X)

# Fit the model
biden_econ_poll = model.fit()

# Print the model summary
print(biden_econ_poll.summary())
```

```python
# Effect of education on Biden's polling
# No significant predictors

X = biden_polls.iloc[:, 7:14]
y = biden_polls.iloc[:, 1:2]

# Define the model
X = sm.add_constant(X)
model = sm.OLS(y, X)

# Fit the model
biden_edu_poll = model.fit()

# Print the model summary
print(biden_edu_poll.summary())
```

```python
# Effect of income on Trump's polling
# No significant predictors

X = trump_polls.iloc[:, 2:7]
y = trump_polls.iloc[:, 1:2]

# Define the model
X = sm.add_constant(X)
model = sm.OLS(y, X)

# Fit the model
trump_econ_poll = model.fit()

# Print the model summary
print(trump_econ_poll.summary())
```

```python
# Effect of income on Biden's donations
# Significant predictors:
    # Median nonfamily income
    # Percent of of people 18 to 64 in poverty
    # Percent of people over 65 in poverty
    # Percent of unemployed civilians

X = biden_donations.iloc[:, 4:9]
y = biden_donations.iloc[:, 3:4]


# Define the model
X = sm.add_constant(X)
model = sm.OLS(y, X)

# Fit the model
biden_econ_donations = model.fit()

# Print the model summary
print(biden_econ_donations.summary())
```
